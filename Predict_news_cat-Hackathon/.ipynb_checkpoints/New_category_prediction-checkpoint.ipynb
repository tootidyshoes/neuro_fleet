### Basic Feature Extraction
<li> Total no. of words</li>

```python
listed = []
for i in range(0,len(train_data['STORY'])):
    listed.append(train_data['STORY'][i].split(' '))
```

```python
listed
```

```python
from collections import Counter
counts = Counter(listed)
print(counts)
```
